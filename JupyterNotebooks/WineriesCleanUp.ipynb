```python
import pandas as pd
```

```python
albaniaDF = pd.read_csv("../data/Wineries/albania.csv") 
albaniaDF
```

```python
australiaDF=pd.read_csv("../data/Wineries/australia.csv")
del australiaDF['latlong']
australiaDF
```

```python
canadaDF=pd.read_csv("../data/Wineries/canada.csv")
del canadaDF['latlong']
canadaDF
```

```python
chileDF=pd.read_csv("../data/Wineries/chile.csv")
del chileDF['latlong']
chileDF
```

```python
new_zealandDF=pd.read_csv("../data/Wineries/new_zealand.csv")
new_zealandDF = new_zealandDF.rename(columns = {'area':'region', 'wine':'web'})
new_zealandDF
```

```python
portugalDF=pd.read_csv("../data/Wineries/portugal.csv")
del portugalDF['latlong']
portugalDF
```

```python
south_africaDF=pd.read_csv("../data/Wineries/south_africa.csv")
del south_africaDF['latlong']
south_africaDF
```

```python
united_statesDF=pd.read_csv("../data/Wineries/united_states.csv")
united_statesDF['country']='United States'
united_statesDF=united_statesDF.rename(columns={'state':'region','web_site':'web','winery_name':'winery'})
united_statesDF
```

```python
wineriesDF=pd.concat([albaniaDF, australiaDF,canadaDF,chileDF,new_zealandDF,portugalDF,south_africaDF,united_statesDF]) 
wineriesDF
```

```python
# saving the DataFrame as a CSV file 
wineriesCSV = wineriesDF.to_csv('../data/wineriesCleaned.csv', index = True) 
```
